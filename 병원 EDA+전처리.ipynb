```python
x_train_mean = np.mean(using[num], axis=0)
x_train_std  = np.std(using[num], axis=0)

# Numpy 브로드캐스팅 기능을 활용해 x_train의 평균과 표준편차로 x_train, x_valid의 스케일링을 진행해줍니다.
x_train.loc[:, num] = (x_train[num] - x_train_mean) / x_train_std 
x_valid.loc[:, num] = (x_valid[num] - x_train_mean) / x_train_std
```
